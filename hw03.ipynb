```julia
using Plots
### Part 1 ###

# Given data
v = 6.0            # velocity, km/day
Kr = 0.55          # reaeration, 1/day
Kc = 0.35          # CBOD decay, 1/day
Kn = 0.25          # NBOD decay, 1/day
DO_sat = 10.0      # saturated DO, mg/L

# Inflow and waste stream characteristics
flow_river = 100000.0  # m3/day
flow_ws1 = 10000.0     # m3/day
flow_ws2 = 15000.0     # m3/day

DO_river = 7.5    # mg/L
DO_ws1 = 5.0      # mg/L
DO_ws2 = 5.0      # mg/L

CBOD_river = 5.0  # mg/L
CBOD_ws1 = 50.0   # mg/L
CBOD_ws2 = 45.0   # mg/L

NBOD_river = 5.0  # mg/L
NBOD_ws1 = 35.0   # mg/L
NBOD_ws2 = 35.0   # mg/L

# Compute the initial mixed concentrations after waste stream 1
flow_combined1 = flow_river + flow_ws1
DO_combined1 = (flow_river * DO_river + flow_ws1 * DO_ws1) / flow_combined1
CBOD_combined1 = (flow_river * CBOD_river + flow_ws1 * CBOD_ws1) / flow_combined1
NBOD_combined1 = (flow_river * NBOD_river + flow_ws1 * NBOD_ws1) / flow_combined1

# Define Streeter-Phelps terms 
function dissolved_oxygen(d, C0, B0, N0, U)
    α1 = exp(-Kr * d / U)
    α2 = (Kc / (Kr - Kc)) * (exp(-Kc * d / U) - exp(-Kr * d / U))
    α3 = (Kn / (Kr - Kn)) * (exp(-Kn * d / U) - exp(-Kr * d / U))
    
    # Steady-state solution from Streeter-Phelps
    C = DO_sat * (1 - α1) + C0 * α1 - B0 * α2 - N0 * α3
    return C
end

# Compute DO concentration from the first waste stream to 50 km downstream
distances = 0:0.1:50  # distance in km
C0 = DO_combined1
B0 = CBOD_combined1
N0 = NBOD_combined1

# Velocity U is the same as before, 6 km/day
DO_values = [dissolved_oxygen(d, C0, B0, N0, v) for d in distances]

# Plot DO downstream
display(plot(distances, DO_values, label="Dissolved Oxygen (mg/L)", xlabel="Distance (km)", ylabel="DO (mg/L)", title="Dissolved Oxygen Downstream from Waste Stream 1", lw=2))

# Find the minimum DO value and distance where it occurs
min_DO, min_dist = minimum(DO_values), distances[argmin(DO_values)]
println("Minimum DO concentration: $min_DO mg/L at $min_dist km")
```

### Problem 2 (20 points)

The simplest climate model involves capturing changes to the Earth’s
energy budget (it is commonly called the *energy balance model*, or
EBM). These changes are also called *radiative forcings* (RF), and can
result from several causes, including greenhouse gas emissions, volcanic
eruptions, and changes to the solar cycle. The EBM treats the Earth as a
0-dimensional sphere covered with water, which absorbs heat in response
to radiative forcings. Chanwith global temperature changes resulting
from imbalances in the average (over the entire surface area) heat flux.

The EBM equations are:

$$
\begin{align*}
\overbrace{\frac{dH}{dt}}^{\text{change in heat}} &= \overbrace{F}^{\substack{\text{radiative} \\ \text{forcing}}} - \overbrace{\lambda T}^{\substack{\text{change in} \\ \text{temperature}}} \\
\underbrace{C}_{\substack{\text{ocean heat} \\ \text{capacity}}} \frac{dT}{dt} &= F - \lambda T \\
c\underbrace{d}_{\substack{\text{ocean} \\ \text{mixing depth}}} \frac{dT}{dt} &= F - \lambda T,
\end{align*}
$$

where $c = 4.184\times 10^6 \mathrm{J/K/m}^2$ is the specific heat of
water per area, $d$ is the depth of the ocean mixed layer (we’ll assume
$d = 86 \mathrm{m}$), and $\lambda$ is the **climate feedback factor**
and controls how much the Earth warms in response to increased radiative
forcing (assume
$\lambda = 2.1^\circ \mathrm{C}/(\mathrm{W}/\mathrm{m}^2$)). The total
radiative forcing $F = F_\text{non-aerosol} + \alpha F_\text{aerosol}$,
where $\alpha$ is an uncertain scaling factor reflecting aerosol-cloud
feedbacks (we’ll assume $\alpha = 0.8$).

The code below loads historical and projected radiative forcings (under
the SSP5-8.5 future emissions scenario, which is the most extreme of the
scenarios used to project climate change impacts) from
`data/ERF_ssp585_1750-2500.csv` into a `DataFrame` object and calculates
the non-aerosol and aerosol components of those forcings.

The continuous EBM equation is given by:

$C \frac{dT}{dt} = F - \lambda T$

Discretizing this using Euler's method over time results in:

$C \frac{T(t + \Delta t) - T(t)}{\Delta t} = F(t) - \lambda T(t)$

$T(t + \Delta t) = T(t) + \frac{\Delta t}{C}(F(t) - \lambda T(t))$

where $T(t)$ is the temperature anomaly at time $t$, $F(t)$ is the radiative forcing at time $t$, and $\lambda$ id the climate feedback factor. 



```julia
using CSV, DataFrames, Plots
### Part 2 ###

# Load the forcing data
forcings_all = CSV.read("data/ERF_ssp585_1750-2500.csv", DataFrame, delim=",")

# Constants
c = 4.184e6  # J/K/m² (specific heat of water per area)
d = 86  # m (ocean mixing depth)
C = c * d  # Ocean heat capacity (J/K/m²)
lambda_mean = 2.1  # Climate feedback factor (expected value)
alpha = 0.8  # Scaling factor for aerosol forcing
dt = 31_558_152  # Time step in seconds (1 year)

# Separate the forcings
forcing_aerosol_rad = forcings_all[!,"aerosol-radiation_interactions"]
forcing_aerosol_cloud = forcings_all[!,"aerosol-cloud_interactions"]
forcing_aerosol = forcing_aerosol_rad + forcing_aerosol_cloud
forcing_total = forcings_all[!,"total"]
forcing_non_aerosol = forcing_total - forcing_aerosol

# Initialize temperature and time arrays
T = zeros(length(forcings_all[!,"year"]))  # Temperature anomalies
years = forcings_all[!,"year"]

# Simulate temperature anomalies using the discretized EBM
for t in 2:length(years)
    F = forcing_non_aerosol[t] + alpha * forcing_aerosol[t]  # Total forcing
    T[t] = T[t-1] + dt/C * (F - lambda_mean * T[t-1])  # Euler step for temperature
end

# Plot the temperature anomalies over time
p_temp = plot(years, T, label="Temperature Anomalies (°C)", xlabel="Year", ylabel="Temperature Anomaly (°C)", linewidth=2)
display(p_temp)
```

```julia
using Distributions, Statistics
### Part 3 ###

# Define the distribution for lambda
lambda_dist = LogNormal(log(2.1), log(2)/4)

n_simulations = 1000  # Number of Monte Carlo simulations
T_mc = zeros(n_simulations)  # Store the temperature in 2100 for each simulation

# Monte Carlo Simulation for lambda
for i in 1:n_simulations
    lambda_sample = rand(lambda_dist)  # Sample lambda from LogNormal distribution
    T_temp = 0.0  # Initial temperature anomaly at T(0)
    for t in 2:length(years)
        F = forcing_non_aerosol[t] + alpha * forcing_aerosol[t]
        T_temp = T_temp + dt/C * (F - lambda_sample * T_temp)  # Euler step for temperature with sampled lambda
    end
    T_mc[i] = T_temp  # Record the temperature anomaly in 2100
end

# Calculate the mean and 95% confidence interval for temperature in 2100
T_mean = mean(T_mc)
T_95_ci = quantile(T_mc, [0.025, 0.975])

# Results
println("Expected Temperature in 2100: $T_mean °C")
println("95% Confidence Interval: ($T_95_ci[1] °C, $T_95_ci[2] °C)")
```

This expected temperature found here approximately close to the iterative simulation for temperature anomolies in part 2 (~6.12). 

The sufficiency of the sample set is based on the convergence of the mean and the confidence interval. As I increased the number of Monte Carlo simulations, I noticed that the mean temperature anomaly and the bounds of the 95% confidence interval began to stabilize. This means that further increasing the number of simulations would likely not result in significant changes to the results (Law of Large Numbers).
